# Determining optimal number of clusters  

```python
from sklearn.cluster import KMeans
WCSS=[]
for i in range(1,11):
    model=KMeans(n_clusters=i, init="k-means++", random_state=2)
    model.fit(X)
    
    WCSS.append(model.inertia_)
```
